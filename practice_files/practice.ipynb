```python
from splinter import Browser
from bs4 import BeautifulSoup as soup
```

```python
# Windows users
executable_path = {'executable_path': 'chromedriver.exe'}
browser = Browser('chrome', r'C:\Users\eblak\Class_Folder\Mission-to-Mars', headless=False)
```

```python
# Visit the Quotes to Scrape site
url = 'http://quotes.toscrape.com/'
browser.visit(url)
```

```python
# Parse the HTML
html = browser.html
html_soup = soup(html, 'html.parser')
```

```python
# Scrape the Title
title = html_soup.find('h2').text
title
```

```python
# Scrape the top ten tags
tag_box = html_soup.find('div', class_='tags-box')
# tag_box
tags = tag_box.find_all('a', class_='tag')

for tag in tags:
    word = tag.text
    print(word)
```

```python
# skills drill Visit the Books site to scrape
url = 'http://books.toscrape.com/'
browser.visit(url)
```

```python
# skills drill parse the HTML
html = browser.html
html_soup = soup(html, 'html.parser')
```

```python
# skills drill scrape the book list
# variable stores results from the parsing
book_box = html_soup.find('div', class_='side_categories')

books = book_box.find_all('ul', class_='nav-list')

for book in books:
    word = book.text
    print(word)
```
